```python
## Creating a list of continuous variables

con_variables = []
for col in df:
    if df[col].nunique() > 12 and col not in ['track', 'artist', 'uri', 'target']:
        con_variables.append(col)
        
con_variables
```

```python
## Creating a list of categorical variables

cat_variables = []
for col in df:
    if df[col].nunique() <= 12 and col != 'target':
        cat_variables.append(col)
        
cat_variables
```

```python
pitch_notation_dict = {0: 'C', 1: 'C-Sharp', 2: 'D', 3: 'D-Sharp', 4:'E', 5:'F', 6:'F-Sharp',
                       7:'G', 8:'G-Sharp', 9:'A', 10:'B-Flat', 11:'B'
    }

df.key = df.key.replace(pitch_notation_dict)
```

```python
mode_dict = {1:'Major', 0:'Minor'}
df['mode'] = df['mode'].replace(mode_dict)
```

```python
hit_data = df.loc[df.target == 1]
flop_data = df.loc[df.target == 0]
```

```python
hit_data.head()
```

```python
flop_data.head()
```

```python
hit_songs = df.loc[df['target'] == 1]
flop_songs = df.loc[df['target'] == 0]
for feat in con_variables:
    sns.distplot(hit_songs[feat], label='Hit Songs' )
    sns.distplot(flop_songs[feat], label='flop Songs')
    plt.title('Song Distribution By Class')
    
    plt.legend()

    plt.show()
```

```python
for feat in cat_variables:
    fig = plt.figure(figsize=(10,8))
    

    hit_data[feat].value_counts().plot(kind='bar', color='b', position=1, width=.4, label='Hit Songs')
    flop_data[feat].value_counts().plot(kind='bar', color='g', position=0, width=.4, label='flop Songs')
    plt.legend(loc=1, )

    plt.ylabel('Number of Songs')
    plt.xlabel(feat)
    plt.title('Hit Songs vs. flop Songs')
    plt.show()
```
